# Missing Value

```python
from scipy.stats import wilcoxon

test_auto = [0.62,0.59,0.65,0.62,0.63,0.62,0.65,0.63,0.6,0.6,0.62,0.61]
test_drop = [0.62,0.61,0.62,0.67,0.53,0.57,0.56,0.65,0.63,0.66,0.67,0.70]
test = wilcoxon(test_auto,test_drop)
test
```

```python
validation_drop_auto = [0.59,0.61,0.6,0.61,0.64,0.64,0.61,0.64,0.62,0.61,0.58,0.6]
validation_drop_drop = [0.62,0.64,0.68,0.64,0.57,0.53,0.58,0.64,0.68,0.66,0.69,0.68]
validation_drop = wilcoxon(validation_drop_auto,validation_drop_drop)
validation_drop
```

```python
validation_fill_auto = [0.57,0.62,0.49,0.63,0.49,0.61,0.5,0.62,0.62,0.62,0.59,0.62]
validation_fill_drop = [0.58,0.58,0.48,0.63,0.5,0.48,0.47,0.6,0.61,0.62,0.62,0.63]
validation_fill = wilcoxon(validation_fill_auto,validation_fill_drop)
validation_fill
```

```python
import matplotlib.pyplot as plt
# make a figure of 3 subplots
fig, ax = plt.subplots(1,3,figsize=(15,5))
# each subplot constains 2 boxplots
ax[0].boxplot([test_auto,test_drop])
ax[1].boxplot([validation_drop_auto,validation_drop_drop])
ax[2].boxplot([validation_fill_auto,validation_fill_drop])
# set the labels
ax[0].set_xticklabels(['Imputation','Drop'])
ax[1].set_xticklabels(['Imputation','Drop'])
ax[2].set_xticklabels(['Imputation','Drop'])

# set the title
ax[0].set_title('Test')
ax[1].set_title('Validation drop')
ax[2].set_title('Validation imputation')
# set the y label
ax[0].set_ylabel('Accuracy')
ax[1].set_ylabel('Accuracy')
ax[2].set_ylabel('Accuracy')
plt.show()
```
